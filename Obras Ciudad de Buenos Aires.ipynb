### Transformaciones del data type

```python
# Convertir la columna a formato datetime (si no lo está)

df["Periodo"] = pd.to_datetime(df["Periodo"], errors="coerce").dt.date
```

```python
# Replace commas with periods and convert to float

df["Long"] = pd.to_numeric(df["Long"].str.replace(",", ".", regex=False), errors="coerce")
df["Lat"] = pd.to_numeric(df["Lat"].str.replace(",", ".", regex=False), errors="coerce")
```

```python
# Remove thousands separators and convert to float

df["MetrosAConstruir"] = df["MetrosAConstruir"].str.replace(".", "", regex=False)
df["MetrosAConstruir"] = df["MetrosAConstruir"].str.replace(",", ".", regex=False)
df["MetrosAConstruir"] = pd.to_numeric(df["MetrosAConstruir"], errors="coerce")
```

```python
# Convert the following columns to date type
# List of columns containing dates
date_columns = ["Fecha_inicio_obra", "Fecha_ultimo_AVO", "Fecha_Estado_Tramite"]

# Loop to convert each column to datetime
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors="coerce", format="%Y/%m/%d")
```

```python
# Convert categorical columns to category data type
categorical_columns = ['Estado', 'EstadoTramite', 'Tipo_Obra', 'Destino', 
                      'Exp_DGROC', 'Calle', 
                      'Ultimo_AVO_Solicitado', 'Profesional', 'Barrios', 
                      'Comuna']

for col in categorical_columns:
    df[col] = df[col].astype('category')

# Check memory usage before and after conversion
print("Memory usage after conversion:")
print(df.memory_usage(deep=True).sum() / (1024 * 1024), "MB")

# Verify the data types were changed
print("\nUpdated data types:")
print(df.dtypes)
```

This code converts not only the specified columns but also other object columns that appear to be categorical in nature. Converting these columns to the 'category' data type can significantly reduce memory usage, especially for columns with repeated values.

```python
# We are going to clean up the amount of procedures so that we can group everything into fewer categories.

def simplificar_estado(estado):
    if estado in ["En AVO 1", "En AVO 2", "En AVO 3", "En AVO 4", "Con Pedido De Avo1", "Con Pedido De Avo2", "Con Pedido De Avo3" ]:
        return "En Verificación"
    elif estado in ["Pedido de AVO 1 autorizado", "Pedido de AVO 2 autorizado", 
                    "Pedido de AVO 3 autorizado", "Pedido de AVO 4 autorizado"]:
        return "Verificación Autorizada"
    elif estado in ["En Avo 1", "En Avo 2", 
                    "En Avo 3", "En Avo 4"]:
        return "En Verificación"
    elif "excavación" in estado.lower():  # Si contiene la palabra "Excavación"
        return "En Excavación"
    elif estado in ["Pedido De Demolición Autorizado", "Pedido De Demolición Final Autorizado", 
                    "Pedido De Demolición Final Observado", "Pedido De Demolición Observado", "Pedido De Demolición Observado Completo"]:
        return "Pedido de Demolición"
    elif "Inicio De Obra" in estado.lower():  # Si contiene la palabra "Excavación"
        return "Inicio De Obra" 
    elif estado in ["Pedido De Avo 1 Autorizado", "Pedido De Avo 1 Observado", 
                    "Pedido De Avo 2 Autorizado", "Pedido De Avo 2 Observado", "Pedido De Avo 3 Autorizado",
                    "Pedido De Avo 3 Observado"]:
        return "Pedido de Verificación"
    elif estado == "En Demolición Final":
        return "En Demolición"
    return estado  # Keeps values that do not need to be modified

# Apply the function to the EstadoTramite column
df["EstadoTramite"] = df["EstadoTramite"].apply(simplificar_estado)
```
